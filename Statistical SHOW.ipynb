```python
# IMPORT STAT TOOLS
from pandas import Series, DataFrame
from os import walk
import pandas as pd
```

```python
# PATHS
input_filepath = 'D:\\Data3\\'

# GETTING ALL FILES AS LIST

all_files = []
for (dirpath, dirnames, filenames) in walk(input_filepath):
    all_files.extend(filenames)
    break

# SELECTING PCAP FILES AS LIST

csv_files = []
for filename in all_files:
    if '.pcap.csv' in filename:
        csv_files.append(filename)
```

```python
main_dframe = DataFrame()
```

```python
index = 0
dframe_list = []

for filename in csv_files:
    index += 1
    print('|', end='')
    if index%80 == 0:
        print()
    
    # READ FILE AND APPEND
    dframe = pd.read_csv(input_filepath + filename, index_col=None, header=0)
    dframe_list.append(dframe)
    
main_dframe = pd.concat(dframe_list)

print('\nFiles processed:', index)
```

```python
# IMPORT STAT TOOLS
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# IMPORT PLOTTING
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# CHANGE FIGURE SIZE
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12

# IMPORT OTHER
import os
```

```python
frame_arr = np.array_split(main_dframe, 9)
```

```python
figsize = (16,8)
cols = 3
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3,cols)
```

```python
fig1 = plt.figure(num=1, figsize=figsize)
ax = []
index = 0
for f in frame_arr:
    print('Frame .' )
    tmp = DataFrame(f)
    row = index // cols
    col = index % cols
    ax.append(fig1.add_subplot(gs[row, col]))
    ax[-1].scatter(tmp['frame.time_epoch'], tmp['ip.ttl'], s=1, marker=',', alpha=0.1)
    index += 1
    
```
