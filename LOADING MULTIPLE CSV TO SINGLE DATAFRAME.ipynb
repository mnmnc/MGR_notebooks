```python
# IMPORT STAT TOOLS
from pandas import Series, DataFrame
from os import walk
import pandas as pd
```

```python
# PATHS
input_filepath = 'D:\\Data\\'

# GETTING ALL FILES AS LIST

all_files = []
for (dirpath, dirnames, filenames) in walk(input_filepath):
    all_files.extend(filenames)
    break

# SELECTING PCAP FILES AS LIST

csv_files = []
for filename in all_files:
    if '.pcap.csv' in filename:
        csv_files.append(filename)
```

#### PROCESSING CSV FILES

```python
main_dframe = DataFrame()
```

```python
# list_ = []
# for file_ in allFiles:
#     df = pd.read_csv(file_,index_col=None, header=0)
#     list_.append(df)
# frame = pd.concat(list_)
index = 0
dframe_list = []
for filename in csv_files:
    index += 1
    print('.', end='')
    if index%100 == 0:
        print()
    dframe = pd.read_csv(input_filepath + filename, index_col=None, header=0)
    dframe_list.append(dframe)
main_dframe = pd.concat(dframe_list)
```

```python
# IMPORT STAT TOOLS
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# IMPORT PLOTTING
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# CHANGE FIGURE SIZE
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12

# IMPORT OTHER
import os
```

```python
def give_date(timestamp):
    import datetime
    return datetime.datetime.fromtimestamp(
            int(timestamp)
        ).strftime('%Y-%m-%d %H:%M:%S')
    
```

```python
plt.scatter(main_dframe['frame.time_epoch'].values,main_dframe['frame.len'].values, s=1, marker=',', alpha=0.1)
plt.title('Frame time vs Frame length', fontsize=18 )
plt.xlabel('Frame time between ' 
           + give_date((main_dframe['frame.time_epoch'].values)[0]) 
           + ' and ' 
           + give_date((main_dframe['frame.time_epoch'].values)[len(main_dframe)-1])
          )
plt.ylabel('Frame length')
```
